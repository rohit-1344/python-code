```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("/content/drive/MyDrive/data2.csv")
print(data)
x=np.array(data[['Roll NO']])
y=np.array(data[['Total']])
print(x)
print(y)
plt.scatter(x,y)
plt.xlabel("ROLL NO")
plt.ylabel("TOTAL")
alpha=0.01
iterations=700
b1=0 #weight1
b0=0 #weight0

for i in range (iterations):
    y_pred=b1*x+b0
    error = y_pred-y

    b1 = b1 - (2*alpha/len(y))*np.sum(error*x)
    b0 = b0 - (2*alpha/len(y))*np.sum(error)

y_predict=b1*x+b0
plt.scatter(x,y)
plt.plot(x,y_predict)
```
