```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('/content/drive/MyDrive/data2.csv')
print(data)
x=np.array(data[['Roll NO']])
print(x)
y=np.array(data[['Total']])
print(y)
x_mean=np.mean(x)
y_mean=np.mean(y)
print(x_mean)
print(y_mean)
plt.xlabel("Roll NO")
plt.ylabel("Total")
plt.scatter(x,y)
```

```python
num=0
den=0
for i in range(len(x)):
   num = num + (x[i] - x_mean)*(y[i] - y_mean)
   den = den + (x[i] - x_mean)**2
b1 = num/den
b0 = (y_mean - b1 * x_mean)
print(b0,b1)
y_pred = b1*x+b0
plt.scatter(x,y,color='red')
plt.plot(x,y_pred)
plt.show()
num_r=0
den_r=0
for i in range(len(x)):
  num_r= num_r + (y[i] - y_pred)**2
  den_r= den_r + (x[i] - x_mean)**2
r_squ=1-(num_r/den_r)
print(r_squ)
```
